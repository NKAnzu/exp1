```python
#파일 읽기
csv_path = os.getenv("HOME") + '/aiffel/dataton/OTT service'

amazon_path = csv_path+'/amazon'
disney_path = csv_path+'/disney'
netflix_path = csv_path+'/netflix'
paramount_path = csv_path+'/paramount'

amazon_title = pd.read_csv(amazon_path+'/titles.csv')
amazon_credit = pd.read_csv(amazon_path+'/credits.csv')

disney_title = pd.read_csv(disney_path+'/titles.csv')
disney_credit = pd.read_csv(disney_path+'/credits.csv')

netflix_title = pd.read_csv(netflix_path+'/titles.csv')
netflix_credit = pd.read_csv(netflix_path+'/credits.csv')

paramount_title = pd.read_csv(paramount_path+'/titles.csv') 
paramount_credit = pd.read_csv(paramount_path+'/credits.csv')
def genre_onehot(datas): # d = dataset
    #데이터 전처리 - 연령 컬럼 제거
    dataset_genres = []
    for d in datas:
        del d['age_certification']
        #
        #데이터 전처리 - null값 제거
        d['genres'].dropna(inplace=True)
    
        dataset_genres.append(d['genres']) # 데이타셋별 genres 컬럼 내용 추가.
    
    #전체 장르 골라내기
    genres=[]
    for d in dataset_genres:
        for i in range(len(d)):
            element = d[i][1:-1]
            for j in element.split(','):
                #genres.append(j.strip()[1:-1])
                genres.append(j.strip())
    genres = pd.unique(genres)
    genres.sort()
    genres = np.delete(genres,0)
    print("장르: " , genres)
    
    #원핫인코딩을 위한 zero_matrix 생성
    #각 데이터셋 별 dummies
    dummies_list=[] #원핫인코딩된 내용이 각 index 0,1,2,3별로 들어가있다.
    for d in datas:
        zero_matrix_data = np.zeros((len(d),len(genres)))
        dummies = pd.DataFrame(zero_matrix_data, columns=genres)
        dummies.columns.get_indexer(genres)
        dummies_list.append(dummies)
    
    dummies_onehot=[]
    for d in range(len(datas)): #int index
        for i, genre in enumerate(datas[d].genres):
            #print(i,genre)
            indices = dummies_list[d].columns.get_indexer(genre[1:-1].split(', '))
            dummies_list[d].iloc[i, indices] = 1 #더미 테이블에 넣기 (원핫인코딩)
            #print(type(indices))
        dummies_onehot.append(dummies_list[d])
    #print(dummies_list[0])
        del dummies_onehot[d]['\'drama\'']
        
    #print(type(dummies_onehot[0]))
    
    
    #col들을 죄다 박는다 (genre내용들임)
    global col_list
    col_list=[]
    for dm in dummies_onehot:
        col=[]
        for i in dm.columns:
            col.append(i[1:-1])
        col = map(lambda x: str.replace(x," ",""),col)
        col = map(lambda x: str.replace(x,"'",""),col)
        col = list(col)
        #col.remove('drama')
        col_list.append(col)
        
    #onehot인코딩된 것들을 기존 데이터셋에 붙인다.
    datas_onehot=[]
    datas_year_genre = []
    for d in range(len(datas)):
        dummies_list[d].columns = col_list[d]
    
    for d in range(len(datas)): #더미 합치기 
        datas_onehot.append(datas[d].join(dummies_list[d]))
        del datas_onehot[d]['genres']
    
    return datas_onehot
```
